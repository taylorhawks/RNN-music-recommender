```python
def get_playlist_position(song,playlist):
    return list(df[df['user_playlistname'] == playlist].artist_trackname == song).index(True)
```

```python
df['playlist_position'] = df.apply(lambda row: get_playlist_position(row['artist_trackname'], row['user_playlistname']), axis=1)
```
